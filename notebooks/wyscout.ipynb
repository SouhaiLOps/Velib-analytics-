```python
from dataclasses import is_dataclass, asdict

meta = dataset.metadata
print(type(dataset))        # -> kloppy.domain.event.EventDataset
print(type(meta))           # -> type de Metadata

# Voir les champs disponibles
if is_dataclass(meta):
    print(asdict(meta).keys())
else:
    print([a for a in dir(meta) if not a.startswith("_")])
```
